```python
sorted(nx.edge_betweenness_centrality(G),reverse=False)
```

```python
NUM_ITERATIONS = 15 #imp part,trial & error,
for i in range(NUM_ITERATIONS):
    edge_betweenness = nx.edge_betweenness_centrality(G).items()
    edge_to_delete = sorted(edge_betweenness, key=lambda pair: -pair[1])[0][0] #neg sign is for dec order,[1 0 0] is for getting the 2nd element,i.e.,edge betweennss
    
    G.remove_edge(*edge_to_delete)
    
    nx.draw(G, with_labels=True, node_color='r')
    plt.title('Step %s\nEdge %s Deleted'%(i, edge_to_delete), fontsize=20)
    
    plt.show()
    
```
